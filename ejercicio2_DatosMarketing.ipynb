### Tratamiento de la columna 'income'
Trate la columna:

* Eliminando el espacio del nombre de la columna. Puede usar la función replace.

*Para eliminar el espacio en el nombre de la columna usando el método "replace" indicamos que reemplace los espacios por algo vacío, es decir, por nada, de forma que los eliminará. Cabe decir que esto se aplicará a todas las columnas, y, si los hubiera, eliminaría todos los espacios en el nombre de todas las columna. En este caso las columnas están escritas en PascalCase o snake_case*

```python
df.columns = df.columns.str.replace(' ', '')
```

*Otra manera de hacerlo para aplicarlo únicamente a esta columna en caso de que quisieramos evitar que afecte a las otra columnas sería utilizar el método 'rename'*

```python
# df = df.rename(columns={' Income ':'Income'})
```

* Quite el símbolo del dolar. Puede usar la función replace.

*Usamos nuevamente el método 'replace' esta vez para eliminar el símbolo del dólar de la columna 'income', indicando que reemplace dicho símbolo por nada.*

```python
df['Income'] = df['Income'].str.replace('$', '')
```

* Cambie la columna al tipo float. Puede usar la función astype(...)

*Para cambiar el tipo de la columna en primer lugar debemos eliminar los puntos y las comas del string, de forma que solamente se quede el número total. Esto lo hacemos nuevamente con 'replace'.*

```python
df['Income'] = df['Income'].str.replace(',', '')
df['Income'] = df['Income'].str.replace('.', '')
```

*Una vez hemos eliminado puntos y comas, podemos utilizar la función 'astype' para convertir la variable en float*

```python
df['Income'] = df['Income'].astype(float)
```

### Rellene los valores nulos
Use la mediana de la columna para rellenar aquellos valores nulos.
La mediana puede ser calculada mediante '.median()'

```python
#  creamos una variable que contenga la mediana de la columna Income, siendo esta la única que contiene valores nulos
mediana = df['Income'].median()

# rellenamos los valores nulos en la columna con la mediana utilizando 'fillna'
df['Income'].fillna(mediana, inplace=True)
```

### Otras tareas de ingeniería de datos que podemos hacer para facilitar la vida al equipo de ventas y marketing

```python
# crear una nueva columna con el número total de menores de edad en el hogar (nueva columna 'Dependents') a partir de la suma de 'Kidhome' y 'Teenhome'.
df['Dependents'] = df['Kidhome'] + df['Teenhome']
```

```python
# crear una nueva columna con el año de cuando el cliente compró por primera vez (nueva columna 'Year_Customer') partir de 'Dt_Customer'.
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])   #en primer lugar nos aseguramos que tenga formato fecha
df['Year_Customer'] = df['Dt_Customer'].dt.year
```

```python
# crear una nueva columna con la cantidad total gastada (nueva columna "TotalMnt") a partir de la suma de todas las columnas que contienen la palabra clave 'Mnt'.
df['TotalMnt'] = df.filter(like='Mnt').sum(axis=1)  #usando el método 'filter'
```

```python
# crear una nueva columna con las compras totales (nueva columna "TotalPurchases") a partir de la suma de todas las columnas que contienen la palabra clave 'Purchases'.

df['TotalPurchases'] = df.filter(like='Purchases').sum(axis=1) 
```

```python
# crear una nueva columna con el número total de campañas aceptadas por clientes (nueva columna 'TotalCampaignsAcc') a partir de la suma de todas las columnas que contienen las palabras clave 'Cmp' y 'Response' (significa que es la última campaña).
columnas = [columna for columna             # identificamos las columnas que contienen "Cmp" y "Response"
            in df.columns 
            if 'Cmp' in columna 
            or 'Response' in columna] 

# y creamos la nueva columnas a partir de la extracción realizada
df['TotalCampaignsAcc'] = df[columnas].sum(axis=1)
```
